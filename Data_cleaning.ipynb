```python
#Lets keep the date and year as the first few columns for easy understanding
df = df[['Date Segment','Year','Date', 'Month'] + [col for col in df if col not in ['Date Segment','Date', 'Month', 'Year']]]
```

```python
#Function to perform the same cleaning on the other datasets
def cleaning_data(data):
    df=data
    df = df.drop('To Date', axis=1)
    
    df['From Date'] = pd.to_datetime(df['From Date'], format='%d-%m-%Y %H:%M')
    #Extract the date, month, and year into separate columns
    df['Date'] = df['From Date'].dt.day
    df['Month'] = df['From Date'].dt.month_name()
    df['Year'] = df['From Date'].dt.year
    
    df['From Date'] = pd.to_datetime(df['From Date'], format='%d-%m-%Y')
    #Drop unnecessary columns
    req_col=['From Date','Date','Month','Year','PM2.5','PM10','NO2','SO2','NH3','CO','Ozone']
    df=df[req_col]
    
    #Create a new column 'Date Segment'
    df['Date Segment'] = 'Other'
    #Oct 2020 - Jan 2021
    df.loc[(df['From Date'] >= '01-10-2020') & (df['From Date'] <= '31-01-2021'), 'Date Segment'] = 'Q1'
    #Oct 2021 - Jan 2022
    df.loc[(df['From Date'] >= '01-10-2021') & (df['From Date'] <= '31-01-2022'), 'Date Segment'] = 'Q2'
    #Oct 2022 - Jan 2023
    df.loc[(df['From Date'] >= '01-10-2022') & (df['From Date'] <= '31-01-2023'), 'Date Segment'] = 'Q3'
    #Lets keep the date and year as the first few columns for easy understanding
    df = df[['Date Segment','Year','Date', 'Month'] + [col for col in df if col not in ['Date Segment','Date', 'Month', 'Year']]]
    
    #Format necessary columns
    df['From Date'] = pd.to_datetime(df['From Date'], format='%d-%m-%Y %H:%M').dt.strftime('%d-%m-%Y')
    
    #Replacing 'None' with 0
    df = df.replace('None', 0)
    
    #for convinience, convert Month to 'string'
    df['Month'] = df['Month'].astype('string')
    return df
```

```python
#Cleaning Punjabi Bagh data
df= pd.read_excel("Combined_PB.xlsx")
df_PB= cleaning_data(df)
df_PB
```

```python
#Cleaning Mundka data
df= pd.read_excel("Combined_MK.xlsx")
df_MK= cleaning_data(df)
df_MK
```

```python
#save the cleaned data
df_AV.to_excel('AV_cleaned.xlsx', index=False)
df_PB.to_excel('PB_cleaned.xlsx', index=False)
df_MK.to_excel('MK_cleaned.xlsx', index=False)
```
